```python
plt.axvline(x=xcord, ymin=0, ymax=1, c='g')
```

```python
x_points = [0.1, 0.3, 0.6, 0.7, 3]
line_color = ['r', 'b', 'g', 'k', 'y']

for xcord, color in zip(x_points, line_color):
    plt.axvline(x=xcord, ymin=0, ymax=1, c = color)
```

```python
plt.axvline(x=2, ymin=0, ymax=1, c = "#0000FF")
```

```python
x_points = [0.1, 0.3, 0.6, 0.7, 3]
line_color = ['#0990f1', 'b', 'g', '#345678', 'y']

for xcord, color in zip(x_points, line_color):
    plt.axvline(x=xcord, ymin=0, ymax=1, c = color)
```

```python
plt.axvline(x=xcord, ymin=0, ymax=1, c = color, label="Yellow Color Line")
plt.legend()
```

```python
x_points = [0.1, 0.3, 0.6, 0.7, 3]
line_color = ['#0990f1', 'b', 'g', '#345678', 'y']
labels = ['One', 'two', 'three', 'four', 'five']

for xcord, color, label in zip(x_points, line_color, labels):
    plt.axvline(x=xcord, ymin=0, ymax=1, c = color, label=label)
    
plt.legend()
plt.show()
```

```python
plt.vlines(x=0.4, ymin=-1.5, ymax=1.5, color="b", label="Blue Line")
plt.legend()
```

```python
x_points = [0.1, 0.3, 0.6]
yminmax  = [(-1, 1), (1, 4), (-5, 4)]
line_color = ['#0990f1', 'b', 'g']
labels = ['One', 'two', 'three']

for xcord, color, label, ymm in zip(x_points, line_color, labels, yminmax):
    plt.vlines(x=xcord, ymin=ymm[0], ymax=ymm[1], color=color, label=label)

plt.hlines(y=2, xmin=-2, xmax=2, color='k', label="Horizontal Line")

plt.legend()
plt.show()
```
